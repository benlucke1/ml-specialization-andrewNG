```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,50)
fig, ax = plt.subplots()
plt.scatter(1,0,marker='x',linewidth=3,color='black',zorder=10)
ax.plot(x,np.log(x),marker='x',color='blue',label='$log(x)$')
ax.plot(x,-np.log(x),marker='o',color='red',label='$-log(x)$')
ax.plot(np.linspace(-1,11),np.zeros(50),'k--')
ax.set_xlim(-0.5,11)
ax.set_xlabel('$x$')
ax.set_ylabel('f(x)')
plt.legend(loc='upper left',prop={'size': 10})
plt.show()
```

We know that the loss function for the logistic model when $y^{(i)} =1$ is of the form $-\log\left(f_{\overrightarrow{w},b}(\overrightarrow{x}^{(i)})\right)$.

## Loss Functions - a closer look
### When $y^{(i)} = 1$

```python
x = np.linspace(0,1,50)
fig, ax = plt.subplots()
ax.plot(x,-np.log(x),marker='o',color='red',label='$-log(x)$')
ax.plot(np.linspace(-1,11),np.zeros(50),'k--')
ax.set_xlim(0.,1.05)
ax.set_ylim(-.1,5.5)
ax.set_xlabel('Prediction from algorithm')
ax.set_ylabel('Loss')
plt.legend(loc='upper left',prop={'size': 10})
plt.title('Loss function when $y^i = 1$')
plt.show()
```

The model will be pushed to predict values of 1 since the loss is the lowest there and the overall goal is to minimize the cost.

### When $y^{(i)} = 0$

We know that the loss function for the logistic model when $y^{(i)} =0$ is of the form $-\log\left(f_{\overrightarrow{w},b}(\overrightarrow{x}^{(i)})\right)$.

```python
x = np.linspace(0,1,50)
fig, ax = plt.subplots()
ax.plot(x,-np.log(1-x),marker='o',color='red',label='$-log(x)$')
ax.plot(np.linspace(-1,11),np.zeros(50),'k--')
ax.set_xlim(0.,1.05)
ax.set_ylim(-.1,5.5)
ax.set_xlabel('Prediction from algorithm')
ax.set_ylabel('Loss')
plt.legend(loc='upper left',prop={'size': 10})
plt.title('Loss function when $y^i = 0$')
plt.show()
```

Using a piecewise cost function yields a convex cost and thus can be opitmized using gradient descent and yield a global minimum for a given feature set, $\vec{x}$ and parameters, $\vec{w},b$.

$$J(\vec{w},b) = \frac{1}{m} \sum_{i=1}^m L(f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})$$
